```python
# Handdle null values
df.groupby('YearsWithCurrManager')['Attrition'].value_counts()
```

```python
# unstack() - converts output into table formate
df.groupby('YearsWithCurrManager')['Attrition'].value_counts().unstack()
```

```python
df.groupby('YearsWithCurrManager')['Attrition'].value_counts().unstack().fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
#dropna=True  → remove null groups 
#dropna=False → include null groups
df.groupby('YearsWithCurrManager', dropna=False)['Attrition'].value_counts()
```

```python
df['YearsWithCurrManager'].fillna(df['YearsWithCurrManager'].median(), inplace=True)
```
